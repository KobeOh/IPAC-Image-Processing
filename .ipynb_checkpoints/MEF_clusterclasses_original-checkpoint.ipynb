```python
startTime = datetime.now()

#Open target file and put ID, X & Y coordinates, and v-band mag (for filtering) into a table 

#Brad's Files
#txt = 'ngc_3627_uvis_working_phot_format_bcw_jan_4_2018.txt'
#txt = 'ngc_4395n_uvis_working_phot_format_bcw_apr_27_2017.txt'
#txt = 'ngc_4242_uvis_working_phot_format_bcw_apr_27_2017.txt' bad file
#txt = 'ngc_3351_uvis_working_phot_format_bcw_apr_27_2017.txt'
#txt = 'ngc_45_uvis_working_phot_format_bcw_apr_27_2017.txt'
#txt = 'ngc_4656_uvis_working_phot_format_bcw_apr_27_2017.txt'
#txt = 'ngc_4449_uvis_working_phot_format_bcw_apr_27_2017.txt'
#txt = 'ngc6744_whitmore_merge_jul_12_2017_full_phot.txt'
#txt = 'hlsp_legus_hst_acs-wfc3_ngc5457-c_multiband_v1_pdagb-mwext-avgapcor.tab'

#Public Files
#txt = 'hlsp_legus_hst_wfc3_ngc1566_multiband_v1_padagb-mwext-avgapcor.tab'
#txt = 'hlsp_legus_hst_acs-wfc3_ngc5194-ngc5195-mosaic_multiband_v1_padagb-mwext-avgapcor.tab'
#txt = 'hlsp_legus_hst_acs-wfc3_ngc628-c_multiband_v1_padagb-mwext-avgapcor.tab'
#txt = 'hlsp_legus_hst_acs-wfc3_ngc628-e_multiband_v1_padagb-mwext-avgapcor.tab'
#txt = 'hlsp_legus_hst_wfc3_ngc1313-e_multiband_v1_padagb-mwext-avgapcor.tab'
#txt = 'hlsp_legus_hst_wfc3_ngc1313-w_multiband_v1_padagb-mwext-avgapcor.tab'
#txt = 'hlsp_legus_hst_wfc3_ngc1433_multiband_v1_padagb-mwext-avgapcor.tab'
#txt = 'hlsp_legus_hst_wfc3_ngc1705_multiband_v1_padagb-mwext-avgapcor.tab'
#txt = 'hlsp_legus_hst_wfc3_ngc3344_multiband_v1_padagb-mwext-avgapcor.tab'
#txt = 'hlsp_legus_hst_acs-wfc3_ngc3738_multiband_v1_padagb-mwext-avgapcor.tab'
#txt = 'hlsp_legus_hst_wfc3_ngc4656_multiband_v1_padagb-mwext-avgapcor.tab'
#txt = 'hlsp_legus_hst_acs-wfc3_ngc4449_multiband_v1_padagb-mwext-avgapcor.tab'
#txt = 'hlsp_legus_hst_acs-wfc3_ngc5253_multiband_v1_padagb-mwext-avgapcor.tab'
txt = 'hlsp_legus_hst_acs-wfc3_ngc5474_multiband_v1_padagb-mwext-avgapcor.tab'
#txt = 'hlsp_legus_hst_wfc3_ngc6503_multiband_v1_padagb-mwext-avgapcor.tab'
#txt = 'hlsp_legus_hst_wfc3_ngc7793-e_multiband_v1_padagb-mwext-avgapcor.tab'
#txt = 'hlsp_legus_hst_acs-wfc3_ngc7793-w_multiband_v1_padagb-mwext-avgapcor.tab'

gal = 'ngc5474'
t = Table.read(txt, format='ascii')

#for ngc5457c, which resets object ID numbers after 9999
#for i in range(len(t[9999::])):
#    t[9999+i]['col1'] = 10000+i

#sort by cluster class (cat) and quality (qual)
cat = 4
qual = 1
cut = t[(t['col34']==cat)]
cut = cut[(cut['col33']>=4)]
```

```python
#Load full images for each band

im = [[gal+'_uvis_f275w_sci.fits','f275w'],[gal+'_uvis_f336w_sci.fits','f336w'],[gal+'_uvis_f438w_sci.fits','f438w'],[gal+'_uvis_f555w_sci.fits','f555w'],[gal+'_uvis_f814w_sci.fits','f814w']]#,[gal+'_uvis_f547m_sci.fits','f547m'],[gal+'_uvis_f657n_sci.fits','f657n']]

#Create .fits cutouts 299x299pix centered around each object
#Check X & Y to see if reversed

stamp_radius = 149

for i in range(0,len(cut)):
    for j in range(0,len(im)):
        obj, x, y = cut['col1'][i], cut['col3'][i], cut['col2'][i]
        obj, x, y = int(round(obj)), int(round(x)), int(round(y))
        image = fits.open(str(im[j][0]))
        image_data = image[0].data
        new_array = image_data[(x-stamp_radius-1):(x+stamp_radius),(y-stamp_radius-1):(y+stamp_radius)]
        new_image = fits.PrimaryHDU(new_array)
        hdulist = fits.HDUList([new_image])
        name = str(gal) + '_' + im[j][1] + '_obj_' + str(obj) + '_class' + str(cat) + '_quality' + str(qual)
        hdulist.writeto(name + '.fits', overwrite = True)
```

```python
#set up MEF-format HDUs for all bands

new_hdul = fits.HDUList()
new_hdul.append(fits.ImageHDU())
new_hdul.append(fits.ImageHDU())
new_hdul.append(fits.ImageHDU())
new_hdul.append(fits.ImageHDU())
new_hdul.append(fits.ImageHDU())
new_hdul.append(fits.ImageHDU())
#new_hdul.append(fits.ImageHDU())
#new_hdul.append(fits.ImageHDU())

new_hdul[0].name = gal
new_hdul[1].name = 'f275w'
new_hdul[2].name = 'f336w'
new_hdul[3].name = 'f438w'
#new_hdul[4].name = 'f547m'
new_hdul[4].name = 'f555w'
#new_hdul[6].name = 'f657n'
new_hdul[5].name = 'f814w'
#new_hdul.info()
```

```python
objects = cut['col1']
#gal = 'ngc5457c'
names = []
errors = []
#highpix = []
#copy .fits data from each band into an MEF for all objects

for i in range(len(objects)):
    hdr = new_hdul[0].header
    hdr['Galaxy']=gal
    hdr['ObjectID']=str(int(round(objects[i])))
    hdr['Class']=str(cat)
    hdr['Quality']=str(qual)
    hdr['x_coord']=cut['col2'][i]
    hdr['y_coord']=cut['col3'][i]
    hdr['RA']=cut['col4'][i]
    hdr['Dec']=cut['col5'][i]
    hdr['m_f275w']=cut['col6'][i]
    hdr['e_f275w']=cut['col7'][i]
    hdr['m_f336w']=cut['col8'][i]
    hdr['e_f336w']=cut['col9'][i]
    hdr['m_f438w']=cut['col10'][i]
    hdr['e_f438w']=cut['col11'][i]
    hdr['m_f606w']=cut['col12'][i]
    hdr['e_f606w']=cut['col13'][i]
    hdr['m_f814w']=cut['col14'][i]
    hdr['e_f814w']=cut['col15'][i]
    hdr['CI_555']=cut['col16'][i]
    hdr['N_filt']=cut['col33'][i]
    
    f275 = fits.open(gal + '_f275w_obj_' + str(int(round(objects[i]))) + '_class' + str(cat) + '_quality' + str(qual) + '.fits')
    new_hdul[1].data = f275[0].data
    if np.min(f275[0].data)==np.max(f275[0].data):
        errors.append((str(int(round(objects[i]))),'f275w'))
        print 'error on ' + str(int(round(objects[i]))) + ', f275w = %.2e' % np.max(f275[0].data)
    #if np.max(f275[0].data)>=500:
    #    highpix.append((str(int(round(objects[i]))),'f275w',np.max(f275[0].data)))
    if (cut['col6'][i] == 66.666):
        print cut['col1'][i], ' has f275 zeroed'
        for j in range(len(new_hdul[1].data)):
            for k in range(len(new_hdul[1].data)):
                new_hdul[1].data[j][k] = 0.

    f336 = fits.open(gal + '_f336w_obj_' + str(int(round(objects[i]))) + '_class' + str(cat) + '_quality' + str(qual) + '.fits')
    new_hdul[2].data = f336[0].data
    if np.min(f336[0].data)==np.max(f336[0].data):
        errors.append((str(int(round(objects[i]))),'f336w'))
        print 'error on ' + str(int(round(objects[i]))) + ', f336w = %.2e' % np.max(f336[0].data)
    #if np.max(f336[0].data)>=500:
    #    highpix.append((str(int(round(objects[i]))),'f336w',np.max(f336[0].data)))
    if (cut['col8'][i] == 66.666):
        print cut['col1'][i], ' has f336 zeroed'
        for j in range(len(new_hdul[2].data)):
            for k in range(len(new_hdul[2].data)):
                new_hdul[2].data[j][k] = 0.

    f438 = fits.open(gal + '_f438w_obj_' + str(int(round(objects[i]))) + '_class' + str(cat) + '_quality' + str(qual) + '.fits')
    new_hdul[3].data = f438[0].data
    if np.min(f438[0].data)==np.max(f438[0].data):
        errors.append((str(int(round(objects[i]))),'f438w'))
        print 'error on ' + str(int(round(objects[i]))) + ', f438w = %.2e' % np.max(f438[0].data)
    #if np.max(f438[0].data)>=500:
    #    highpix.append((str(int(round(objects[i]))),'f438w',np.max(f438[0].data)))
    if (cut['col10'][i] == 66.666):
        print cut['col1'][i], ' has f435 zeroed'
        for j in range(len(new_hdul[3].data)):
            for k in range(len(new_hdul[3].data)):
                new_hdul[3].data[j][k] = 0.

    #f547 = fits.open(gal + '_f547m_obj_' + str(int(round(objects[i]))) + '.fits')
    #new_hdul[4].data = f547[0].data
    
    f555 = fits.open(gal + '_f555w_obj_' + str(int(round(objects[i]))) + '_class' + str(cat) + '_quality' + str(qual) + '.fits')
    new_hdul[4].data = f555[0].data
    if np.min(f555[0].data)==np.max(f555[0].data):
        errors.append((str(int(round(objects[i]))),'f555w'))
        print 'error on ' + str(int(round(objects[i]))) + ', f555w = %.2e' % np.max(f555[0].data)
    #if np.max(f555[0].data)>=500:
    #    highpix.append((str(int(round(objects[i]))),'f555w',np.max(f555[0].data)))
    if (cut['col12'][i] == 66.666):
        print cut['col1'][i], ' has f555 zeroed'
        for j in range(len(new_hdul[4].data)):
            for k in range(len(new_hdul[4].data)):
                new_hdul[4].data[j][k] = 0.

    #f657 = fits.open(gal + '_f657n_obj_' + str(int(round(objects[i]))) + '.fits')
    #new_hdul[6].data = f657[0].data
   
    f814 = fits.open(gal + '_f814w_obj_' + str(int(round(objects[i]))) + '_class' + str(cat) + '_quality' + str(qual) + '.fits')
    new_hdul[5].data = f814[0].data
    if np.min(f814[0].data)==np.max(f814[0].data):
        errors.append((str(int(round(objects[i]))),'f814w'))
        print 'error on ' + str(int(round(objects[i]))) + ', f814w = %.2e' % np.max(f814[0].data)
    #if np.max(f814[0].data)>=500:
    #    highpix.append((str(int(round(objects[i]))),'f814w',np.max(f814[0].data)))
    if (cut['col14'][i] == 66.666):
        print cut['col1'][i], ' has f814 zeroed'
        for j in range(len(new_hdul[5].data)):
            for k in range(len(new_hdul[5].data)):
                new_hdul[5].data[j][k] = 0.

    #Save MEF & copy names for .tar use
    
    name = 'MEF_' + gal + '_obj_' + str(int(round(objects[i]))) + '_class' + str(cat) + '_quality' + str(qual) +'.fits'
    names.append(name)
    new_hdul.writeto(name, overwrite = True)
    
    #remove individual .fits files after each MEF is saved
    
    os.remove(gal + '_f275w_obj_' + str(int(round(objects[i]))) + '_class' + str(cat) + '_quality' + str(qual) + '.fits')
    os.remove(gal + '_f336w_obj_' + str(int(round(objects[i]))) + '_class' + str(cat) + '_quality' + str(qual) + '.fits')
    os.remove(gal + '_f438w_obj_' + str(int(round(objects[i]))) + '_class' + str(cat) + '_quality' + str(qual) + '.fits')
    os.remove(gal + '_f555w_obj_' + str(int(round(objects[i]))) + '_class' + str(cat) + '_quality' + str(qual) + '.fits')
    os.remove(gal + '_f814w_obj_' + str(int(round(objects[i]))) + '_class' + str(cat) + '_quality' + str(qual) + '.fits')
```

```python
#Put MEFs into a .tar file

if errors != []:
    errors_table = Table(rows=errors, names=('ID', 'Filter'))
    ascii.write([errors_table['ID'], errors_table['Filter']],gal + '_class' + str(cat) + '_quality' + str(qual) + '_errors.tab', names = ['ID','Filter'], overwrite='True')
#if highpix != []:
#    highpix_table = Table(rows=highpix, names=('ID', 'Filter', 'Pix_Value'))
#    ascii.write([highpix_table['ID'], highpix_table['Filter'], highpix_table['Pix_Value']],gal + '_class' + str(cat) + '_quality' + str(qual) + '_highpix.tab', names = ['ID','Filter', 'Pix_Value'], overwrite=True)

tar = tarfile.open(gal + '_class' + str(cat) + '_quality' + str(qual) + ".tar.gz", "w:gz")
# '_qual-' + str(qual) +
for i in names:
    tar.add(i)
    os.remove(i)
tar.close()

print datetime.now() - startTime
```
